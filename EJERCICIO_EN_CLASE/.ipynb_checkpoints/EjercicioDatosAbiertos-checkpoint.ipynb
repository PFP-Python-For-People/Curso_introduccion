### ¿Qué está pasando y qué implicaciones tiene?

La columa debería ser numérica pero es object, es un área entonces ¿qué implica? que no puede operarse...

```python
# df["area"].mean()

"""
ERROR: Could not convert 56,026.256,026.256,026.256,026.256,026.256,026.256,026.256,026.256,
026.256,026.256,026.256,026.256,026.256,026.256,026.256,026.256,026.256,026.256,026.256,
026.256,026.256,026.256,026.256,026.256,026.256,026.256,026.256,166.756,166.756,166.7 to numeric
"""
```

#### ¿Qué hacemos entonces?

1. Identificar el problema, en este caso: problemas en el separador decimal.
2. Proponer una solución

```python
# En ocasiones las cifras, decimales y sus separadores no funcionan correctamente.... ¿cómo podemos abordarlo?
# Remover la coma de los valores en la serie
df["area"] = df["area"].str.replace(',', '')
df["area"] = df["area"].astype(float)
df.info()
#### Si esto se llama varias veces se daña, aprendemos de excepciones más adelante en el curso.
```

```python
df["area"].mean()
```

#### Tras la tranformación de filas y columnas (transposición) los años quedaron en los índices, ¿cómo los obtenemos?

```python
df["año"] = df.index
df["año"] = df["año"].astype(int)
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt

# Mostrar una línea utilizando plot()
df.plot(x='año', y='area', color='blue', linestyle='--', figsize = (16, 4))

plt.xlim(2012, 2016)
# Configurar etiquetas y título
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de línea')

# Mostrar el gráfico
plt.show()
```

### ¿Cómo podemos mostrar la comparación global del promedio por zona_geo?

```python
grupo = df.groupby("zona_geo").agg({"promedio":"mean"})
grupo
```

### ¿Cómo podemos mostrar la comparación global del promedio por zona_geo por año?

```python
grupo = df.groupby(["zona_geo", "año"]).agg({"promedio":"mean"})
grupo
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear el DataFrame 'grupo' con la agrupación y cálculo del promedio
grupo = df.groupby(["zona_geo", "año"]).agg({"promedio":"mean"})

# Convertir el índice del DataFrame en columnas
grupo = grupo.reset_index()

# Crear una figura y un conjunto de ejes
fig, ax = plt.subplots()

# Iterar sobre los grupos únicos de 'zona_geo'
for zona_geo in grupo['zona_geo'].unique():
    # Filtrar los datos para la zona_geo actual
    datos_zona = grupo[grupo['zona_geo'] == zona_geo]
    
    # Crear la línea para la zona_geo actual
    ax.plot(datos_zona['año'], datos_zona['promedio'], label=zona_geo)

# Configurar etiquetas y título del gráfico
ax.set_xlabel('Año')
ax.set_ylabel('Promedio')
ax.set_title('Gráfico de Líneas por Zona Geográfica')

# Mostrar leyenda
ax.legend()

# Mostrar el gráfico
plt.show()
```
