### 2. Gráficos básicos
[[enlace]](#A00)

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Crear el gráfico de línea
plt.plot(x, y)

# Agregar etiquetas a los ejes x e y
plt.xlabel('X')
plt.ylabel('Y')

# Agregar un título al gráfico
plt.title('Gráfico de línea')

# Mostrar el gráfico
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.scatter(x, y)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de dispersión')

plt.show()
```

```python
x = ['A', 'B', 'C', 'D', 'E']
y = [10, 15, 7, 12, 9]

plt.bar(x, y)

plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.title('Gráfico de barras')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios
data = np.random.randn(1000)

plt.hist(data, bins=50)

plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma')

plt.show()
```

### 3. Personalización de gráficos en Matplotlib
[[enlace]](#A00)

```python
# Personalización de títulos y etiquetas:
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)

plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de línea personalizado')

plt.show()
```

```python
# Cambio de colores y estilos de línea

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, 'g^-', linewidth=1, markersize=10)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de línea con estilo personalizado')
plt.show()
```

```python
# Agregando leyendas:

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

plt.plot(x, y1, label='Línea 1')
plt.plot(x, y2, label='Línea 2')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico con leyendas')
plt.legend()
plt.show()
```

```python
# Cambio de colores utilizando mapas de colores (colormaps):

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, color='Firebrick')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de línea con colormap')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.figure(figsize=(15, 4))  # Cambia el tamaño de la figura
plt.plot(x, y)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de línea con tamaño de figura personalizado')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel('Eje X', fontsize=12)  # Cambia el tamaño de la fuente de las etiquetas
plt.ylabel('Eje Y', fontsize=12)
plt.title('Gráfico de línea con ejes personalizados', fontsize=14)  # Cambia el tamaño de la fuente del título
plt.xticks([1, 2, 3, 4, 5], ['A', 'B', 'C', 'D', 'E'])  # Cambia las etiquetas del eje X
plt.yticks(fontsize=10)  # Cambia el tamaño de la fuente del eje Y
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.style.use('ggplot')  # Aplica el estilo 'ggplot' a los gráficos

plt.plot(x, y)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de línea con estilo predefinido')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

colors = sns.color_palette('husl', len(x))  # Genera una paleta de colores personalizada
plt.scatter(x, y, color=colors)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de dispersión con colores personalizados')
plt.show()
```

### 4. Ejemplos de contexto
[[enlace]](#A00)

```python
import matplotlib.pyplot as plt

meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun']
ventas = [1200, 1500, 1800, 1300, 1600, 2000]

plt.plot(meses, ventas)
plt.xlabel('Mes')
plt.ylabel('Ventas (USD)')
plt.title('Ventas Mensuales')
plt.grid(True)  # Agrega una cuadrícula al gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

temperatura = [25, 28, 30, 32, 27, 29]
humedad = [60, 55, 58, 62, 57, 59]

plt.scatter(temperatura, humedad)
plt.xlabel('Temperatura (°C)')
plt.ylabel('Humedad (%)')
plt.title('Relación entre Temperatura y Humedad')
plt.axhline(60, color='r', linestyle=':')  # Agrega una línea horizontal en y=60
plt.axvline(28, color='g', linestyle=':')  # Agrega una línea vertical en x=28
plt.show()
```

```python
import matplotlib.pyplot as plt

categorias = ['Electrónica', 'Ropa', 'Hogar', 'Deportes']
ventas = [2500, 1800, 3000, 2200]

plt.bar(categorias, ventas)
plt.xlabel('Categoría')
plt.ylabel('Ventas (USD)')
plt.title('Ventas por Categoría')
plt.xticks(rotation=45)  # Rota las etiquetas del eje x para mayor legibilidad
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

edad = np.random.normal(30, 5, 1000)  # Genera una muestra de edad con distribución normal

plt.hist(edad, bins=20)
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Distribución de Edad')
plt.grid(True)  # Agrega una cuadrícula al gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generar datos de muestra
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)
z = np.random.randn(100)
colors = np.random.rand(100)
sizes = 100 * np.random.rand(100)

# Crear la figura y el eje tridimensional
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar los puntos en el espacio tridimensional
ax.scatter(x, y, z, c=colors, s=sizes, alpha=0.7)

# Configurar etiquetas y título
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Gráfico de Dispersión 3D')

# Mostrar el gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de muestra
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Añadir una columna de unos a X para el término de sesgo
X = np.column_stack((np.ones_like(X), X))

# Calcular los coeficientes de la regresión lineal
coefficients = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

# Calcular los valores predichos
y_pred = X.dot(coefficients)

# Graficar los datos de muestra y la línea de regresión
plt.scatter(X[:, 1], y, label='Datos de muestra')
plt.plot(X[:, 1], y_pred, color='r', label='Regresión lineal')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Regresión Lineal')
plt.legend()
plt.show()

# Imprimir los coeficientes de la regresión
print('Coeficientes:', coefficients)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de muestra
np.random.seed(0)
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)
sizes = 100 * np.random.rand(100)

# Crear el gráfico de dispersión
fig, ax = plt.subplots()
scatter = ax.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='plasma')

# Agregar barra de color
cbar = plt.colorbar(scatter)
cbar.set_label('Valores')

# Configurar etiquetas y título
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Gráfico de Dispersión con Colores y Tamaños Variables')

# Mostrar el gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de contexto: Población y Producto Interno Bruto (PIB) de países seleccionados
paises = ['Estados Unidos', 'China', 'Japón', 'Alemania', 'Reino Unido']
poblacion = [328.2, 1393.8, 126.5, 83.2, 66.7]  # en millones
pib = [21427, 15099, 5080, 3846, 2824]  # en miles de millones de dólares

# Crear el gráfico de barras
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax2.bar(paises, pib, label='PIB', color='lightcoral', alpha=0.5)
ax1.bar(paises, poblacion, label='Población', color='skyblue', alpha=0.7)


# Configurar etiquetas y título
ax1.set_xlabel('Países')
ax1.set_ylabel('Población (millones)')
ax2.set_ylabel('PIB (miles de millones de dólares)')

ax1.set_title('Población y PIB de Países Seleccionados')

# Mostrar leyenda
ax1.legend(loc='right')
ax2.legend(loc='upper right')

# Ajustar el espaciado entre las barras
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

years = [1900, 1920, 1940, 1960, 1980, 2000, 2020]
temperature = [14.8, 15.2, 15.8, 16.5, 17.2, 18.1, 18.9]
population = [1600, 1860, 2200, 3030, 4450, 6100, 7800]
industrialization = [False, False, False, True, True, True, True]

# Crear el gráfico
plt.plot(years, temperature, label='Temperatura')
plt.plot(years, population, label='Población')

# Agregar hitos de industrialización con indicadores numéricos
for i in range(len(years)):
    if industrialization[i]:
        plt.scatter(years[i], temperature[i], color='b', marker='o')
        plt.scatter(years[i], population[i], color='g', marker='o')
        plt.annotate(f'#{i+1}', (years[i], temperature[i]), textcoords='offset points', xytext=(0,10), ha='center')
        plt.annotate(f'#{i+1}', (years[i], population[i]), textcoords='offset points', xytext=(0,10), ha='center')

# Agregar etiquetas a los ejes x e y
plt.xlabel('Año')
plt.ylabel('Valor')

# Agregar título al gráfico
plt.title('Crecimiento de la temperatura y población mundial')

# Agregar leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
```
